## Condensed Math

### States

$$
\sigma = \{Red, Blue, Green, ...\}
$$

$$
\epsilon = (\epsilon_1, \epsilon_2, ..., \epsilon_N), \qquad \epsilon_j \in \{-1, +1\}
$$

$$
\tilde{\sigma} = (\sigma, \epsilon)
$$

We aim to match the stationary distribution, which is the **Boltzmann Distribution**. The stationary distribution is determined only by $\sigma$ and is independent of $\epsilon$.

$$
\pi(\sigma) = \frac{e^{-\beta H(\sigma)}}{Z}
$$

This is our energy function. It also does not depend on $\epsilon$.
$$
H(\sigma) = \#Conflicts_{\sigma}
$$

For the full lifted state, directions are auxiliary and uniformly weighted:
$$
\tilde{\pi}(\sigma, \epsilon) = \frac{\pi(\sigma)}{2^N}.
$$

# Transition Logic

Update $j$ changes only the $j$ th color component and uses $\epsilon_j$ to choose the proposed move:
$$
\sigma' = S_j^{\epsilon_j}(\sigma).
$$

The accepted transition is
$$
P\left((S_j^{\epsilon_j}(\sigma),\epsilon) \mid (\sigma,\epsilon)\right) = T_j(\sigma,\epsilon) = q_j A_j(\sigma,\epsilon).
$$

$q_j$ is the proposal probability, which is a constant since all proposals are symmetrical

$A_j(\sigma, \epsilon)$ is the acceptance probability and it is detailed further in metropolis


### Direction Flipping Logic
If the proposal is rejected, only the $j$ th direction flips:
$$
F_j\epsilon = (\epsilon_1, ..., -\epsilon_j, ..., \epsilon_N),
$$
$$
P\left((\sigma,F_j\epsilon) \mid (\sigma,\epsilon)\right) = q_j\bigl(1-A_j(\sigma,\epsilon)\bigr).
$$

All other $\sigma'$ have probability $0$. If different values of $j$ lead to the same destination, their probabilities are added.

### Skewed Detailed Balance

For $\sigma'=S_j^{\epsilon_j}(\sigma)$, the corresponding reverse-direction proposal must undo the move:
$$
S_j^{-\epsilon_j}(\sigma') = \sigma.
$$

The paired accepted flows obey
$$
\tilde{\pi}(\sigma,\epsilon)T_j(\sigma,\epsilon) = \tilde{\pi}(\sigma',F_j\epsilon)T_j(\sigma',F_j\epsilon).
$$

Thus a move from $(\sigma,\epsilon)$ to $(\sigma',\epsilon)$ is paired with a move from $(\sigma',F_j\epsilon)$ to $(\sigma,F_j\epsilon)$, rather than with the same-direction reverse move.

# Metropolis

$$
T_j(\sigma,\epsilon) = q_j A_j(\sigma,\epsilon).
$$

$q_j$ is the probability of selecting node $j$ for an update. With uniform node selection, $q_j=1/N$, and in general
$$
\sum_j q_j=1.
$$

$A_j(\sigma,\epsilon)$ is the acceptance probability for the proposed move $\sigma'=S_j^{\epsilon_j}(\sigma)$:
$$
A_j(\sigma,\epsilon) = \min\bigl(1,e^{-\beta\Delta H_j}\bigr),
$$
where
$$
\Delta H_j=H(\sigma')-H(\sigma).
$$

# Lifted

Direction flips occur after rejections. The probability of the $j$ th direction-flip transition is
$$
L_j(\sigma,\epsilon)=q_j\bigl(1-A_j(\sigma,\epsilon)\bigr).
$$

This auxiliary update changes $\epsilon$ but not $\sigma$, so it does not change the target Boltzmann distribution over colors.

# Boltzmann Proof

$$
\pi(\sigma)=\frac{e^{-\beta H(\sigma)}}{Z}, \qquad \tilde{\pi}(\sigma,\epsilon)=\frac{\pi(\sigma)}{2^N}.
$$

We prove that $\tilde{\pi}$ is stationary for the full lifted state $(\sigma,\epsilon)$. Its color marginal will then be the Boltzmann distribution $\pi(\sigma)$.

### Inflow

Fix a destination lifted state $y=(\sigma,\epsilon)$. For each node $j$, let $\rho_j$ be the unique predecessor color configuration satisfying
$$
S_j^{\epsilon_j}(\rho_j)=\sigma.
$$
Equivalently, $\rho_j=S_j^{-\epsilon_j}(\sigma)$.

There are two ways to enter $(\sigma,\epsilon)$ for each $j$:

- Accept a color proposal from $(\rho_j,\epsilon)$.
- Reject a proposal from $(\sigma,F_j\epsilon)$, which flips its $j$ th direction into $\epsilon$.

Therefore the total incoming probability is
$$
\begin{aligned}
I(\sigma,\epsilon)
&=\sum_j \tilde{\pi}(\rho_j,\epsilon)q_jA_j(\rho_j,\epsilon)\\
&\quad+\sum_j \tilde{\pi}(\sigma,F_j\epsilon)q_j\bigl(1-A_j(\sigma,F_j\epsilon)\bigr).
\end{aligned}
$$

The Metropolis rule and the inverse-proposal property give
$$
\begin{aligned}
\tilde{\pi}(\sigma',\epsilon)q_jA_j(\sigma',\epsilon)
&=\tilde{\pi}(\sigma,F_j\epsilon)q_jA_j(\sigma,F_j\epsilon).
\end{aligned}
$$
This is skewed detailed balance for the paired accepted transitions. Substitute it into the inflow:
$$
\begin{aligned}
I(\sigma,\epsilon)
&=\sum_j\tilde{\pi}(\sigma,F_j\epsilon)q_jA_j(\sigma,F_j\epsilon)\\
&\quad+\sum_j\tilde{\pi}(\sigma,F_j\epsilon)q_j\bigl(1-A_j(\sigma,F_j\epsilon)\bigr)\\
&=\sum_j\tilde{\pi}(\sigma,F_j\epsilon)q_jA_j(\sigma,F_j\epsilon) + \sum_j\tilde{\pi}(\sigma,F_j\epsilon)q_j-\sum_j\tilde{\pi}(\sigma,F_j\epsilon)q_j A_j(\sigma,F_j\epsilon)\\
&=\sum_j\tilde{\pi}(\sigma,F_j\epsilon)q_j\\
&=\sum_j\tilde{\pi}(\sigma,\epsilon)q_j\\
&=\tilde{\pi}(\sigma,\epsilon).
\end{aligned}
$$

The fourth line uses the fact that $\tilde{\pi}$ is uniform over directions, so $\tilde{\pi}(\sigma,F_j\epsilon)=\tilde{\pi}(\sigma,\epsilon)$ for every $j$.

### Outflow and Conclusion

The accepted transitions alone do not sum to $1$, because a proposal can be rejected. Including both outcomes, the total next-step probability is
$$
\begin{aligned}
\sum_y P(y\mid\sigma,\epsilon)
&=\sum_j\left[q_jA_j(\sigma,\epsilon)+q_j\bigl(1-A_j(\sigma,\epsilon)\bigr)\right]\\
&=\sum_jq_j=1.
\end{aligned}
$$

Thus the outgoing probability flow is $\tilde{\pi}(\sigma,\epsilon)$. Since the inflow is also $\tilde{\pi}(\sigma,\epsilon)$, the lifted distribution is stationary. Finally,
$$
\sum_{\epsilon\in\{-1,+1\}^N}\tilde{\pi}(\sigma,\epsilon)
=\sum_{\epsilon\in\{-1,+1\}^N}\frac{\pi(\sigma)}{2^N}
=\pi(\sigma).
$$
Therefore the color configuration follows the Boltzmann distribution.
