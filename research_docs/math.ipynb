# Markov chain States for Graph Coloring

**Graph State**
$$\sigma = \{Red, Green, Blue, Red, ...\}$$

**Direction**
$$\epsilon \in \{+1, -1\}$$

# Markov Transition Logic

This is the probability of seeing the system in that configuration
$$\pi(\sigma)$$

In our model it is calculated using the Boltzman Distribution
$$\pi(\sigma) = \frac{e^{-\beta H(\sigma)}}{Z}$$

$Z$ is the **normalization factor** (the sum of evergy over all states)

This is our energy function
$$H(\sigma) =  \#Conflicts_{\sigma}$$

Ideally, the markov chain should model the energy distribution given by the boltzman distribution


# Transition Logic with Lifting

**Transition probability** of moving from $\sigma$ to $\sigma'$
$$T(\sigma'|\sigma)$$

**Ordinary Detailed Balance**
$$T(\sigma'|\sigma)\pi(\sigma) = T(\sigma|\sigma')\pi(\sigma')$$
$$T(\sigma' \leftrightarrow \sigma)$$
so there is an equal probability for going forward and backwards

**Skewed Detailed Balance**
$$T(\sigma', \epsilon|\sigma, \epsilon)\pi(\sigma) = T(\sigma, -\epsilon|\sigma', -\epsilon)\pi(\sigma')$$

Therefore instead of a reversible flow, we have
$$(\sigma, +) \rightarrow (\sigma', +)$$
which is equal to
$$(\sigma', -) \rightarrow (\sigma, -)$$
which allows us to preserve the correct distribution

# Reversible Gibbs Sampling
$$
T(\sigma'| \sigma) = Q(\sigma'| \sigma)A(\sigma' | \sigma)
$$


$Q$ proposes a candidate state

$A_{\epsilon}$ is the probability of whether to accept it

In Gibbs, all candidate states are accepted so $A(\sigma'| \sigma)$ = 1

$$
T(\sigma'| \sigma) = Q(\sigma'| \sigma)
$$
$$
Q(\sigma' | \sigma) = \frac{e ^ {-\beta H(\sigma')}}{Z}, Z = \sum_{\sigma' \in N(\sigma)}e ^ {-\beta H(\sigma')}
$$

# Metropolis Sampling with Irreversibility
$$T_{\epsilon}(\sigma'| \sigma) = Q(\sigma'|\sigma)A_{\epsilon}(\sigma'| \sigma)$$

$Q$ proposes a candidate state

$A_{\epsilon}$ is the probability of whether to accept it

To calculate the acceptance probability we first equate the forward flow  to the backward flow
$$\pi(\sigma)T(\sigma', \epsilon|\sigma, \epsilon) = \pi(\sigma')T(\sigma, -\epsilon|\sigma', -\epsilon)$$

$$
\pi(\sigma)Q(\sigma'|\sigma)A_{\epsilon}(\sigma'|\sigma) = \pi(\sigma')Q(\sigma|\sigma')A_{\epsilon}(\sigma|\sigma')$$
$$\frac{A_{\epsilon}(\sigma'|\sigma)}{A_{\epsilon}(\sigma|\sigma')} = \frac{\pi(\sigma')Q(\sigma|\sigma')}{\pi(\sigma)Q(\sigma'|\sigma)}
$$
$$
R = \frac{\pi(\sigma')Q(\sigma|\sigma')}{\pi(\sigma)Q(\sigma'|\sigma)}
$$
$$
A_{\epsilon}(\sigma'|\sigma) = min(R, 1)
$$

since $\pi(\sigma) = \frac{e^{-\beta H(\sigma)}}{Z}$, 
$$
\frac{\pi(\sigma')}{\pi(\sigma)}= \frac{e^{-\beta H(\sigma')}}{e^{-\beta H(\sigma)}}= e^{-\beta \Delta H}, \Delta H = H(\sigma') - H(\sigma)
$$

and since it is a uniform distribution which proposes a change at random 
$$
Q(\sigma' | \sigma) = Q(\sigma | \sigma') 
$$
so 
$$
\frac{Q(\sigma' | \sigma)}{Q(\sigma | \sigma')} = 1
$$
so it can be ignored

so the final accceptance probability is

$$
A_{\epsilon}(\sigma'|\sigma) = min(e^{-\beta \Delta H}, 1)
$$


# Lifting

This is a function which returns the flip rate for switching $(\sigma, \epsilon) \rightarrow (\sigma, -\epsilon)$
$$\lambda(\sigma, \epsilon) $$

The Condition for this function is 
$$\lambda(\sigma, \epsilon) - \lambda(\sigma, -\epsilon) = \sum_{\sigma' \in N(\sigma)}{T(\sigma', -\epsilon|\sigma, -\epsilon)-  T(\sigma', \epsilon|\sigma, \epsilon)}$$
where $N(\sigma)$ is the neighbours of the state

This is basically saying: the **difference** in flip rate is determined by the **net difference in the total Transition Rate** between wanting to **go forward** vs wanting to **go backwards** 

From this we can derive that
$$\lambda(\sigma, \epsilon) = max[0, \sum_{\sigma' \in N(\sigma)}{T(\sigma', -\epsilon|\sigma, -\epsilon)-  T(\sigma', \epsilon|\sigma, \epsilon)}]$$

# Magnetization for Ising Model
This is a scalar value which aims to represent the **average alignment of colors of a state**. We have not worked out how to represent and prove that an antiferromagnetic potts model is better mathematically yet

$$
m(\sigma) = \frac{1}{N}\sum^N_{j=1}{\sigma_j}
$$
where $\sigma_j$ is the spin $\in \{-1, 1\}$

### Magnetization Decay

Additionally, the magnetization should behave like
$$
E[m(t)] = E[m(0)]e^{-\frac{t}{\tau}}
$$

And in the Ising model

Additionally, we can prove that 
$$
\tau_{irr} = \frac{\tau_{rev}}{2}
$$
which shows that the irreversible chain should converge twice as fast

# Autocorrelation
This allows us to determine how long the model is able to "remember" its past state to avoid repeating it
$$
Cov(m(0), m(t)) = E[m(0)m(t)] - E[m(0)]E[m(t)]
$$
$$
Cov(m(0), m(t)) = {E[m(0)m(t)] - E[m]^2}
$$

$$
Corr(t) = \frac{Cov(X, Y)}{\sqrt{Var(X)Var(Y)}}
$$

since $Var(m(0))$ is the same as $Var(m(t))$, the bottom term just becomes Var(m)

$$
C(t) = \frac{Cov(m(0), m(t))}{Var(m)}
$$
$$
C(t) = \frac{{E[m(0)m(t)] - E[m]^2}}{E[m^2] - E[m]^2}
$$

To estimate the expectations we use a monte carlo.



# Glauber Mechanics

Proving that irreversibility is better for ising models

to do this, we want to solve for $\tau$ in the equation for magnetic relaxation speed
$$
E[m(t)] = E[m(0)]e^{-\frac{t}{\tau}}
$$
a smaller $\tau$ would mean that the model converges faster

This is from the paper(https://arxiv.org/abs/1301.4801) by Yuji Sakai and Koji Hukushima


**New Terms**
- **J** is the interaction strength
- $\beta = 1 / T$ 
- **$F_j \sigma$** is the configuration obtained by flipping spin j. This is equal to $\sigma'$
- $\gamma = tanh(2\beta J)$
- $\alpha > 0$ is the update rate
- $-1 \geq \delta \leq 1$
- N is the size of the system

More convenient notation for balance

$$
T(\sigma', \epsilon|\sigma, \epsilon)\pi(\sigma) = T(\sigma, -\epsilon|\sigma', -\epsilon)\pi(\sigma') \rightarrow
w_j(\sigma, \epsilon)\pi(\sigma) = w_j(F_j\sigma, -\epsilon)\pi(F_j\sigma)
$$
Where $w_j(\sigma, \epsilon) = T(\sigma', \epsilon|\sigma, \epsilon)$

For SH1 transition probability
$$
w_j(\sigma, \epsilon) = \frac{1}{2}\alpha[1 - \frac{1}{2}\gamma\sigma_j(\sigma_{j-1} + \sigma_{j + 1})](1 - \delta\epsilon\sigma_j)
$$
flip rate is chosen by
$$
\lambda(\sigma, \epsilon) = \sum_j w_j(\sigma, -\epsilon)
$$

According to SH1 type, this is also determined by
$$
\lambda(\sigma, \epsilon) = \frac{1}{2}\alpha N [1 - \gamma u(\sigma) + \delta (1 - \gamma) \epsilon m(\sigma)]
$$

Master Equation
$$
\frac{d}{dt}\pi(\sigma, \epsilon, t) = -[\sum_j w_j(\sigma, \epsilon)] \pi(\sigma, \epsilon, t) + \sum_j w_j(F_j\sigma, \epsilon)\pi(F_j\sigma, \epsilon, t) - \lambda(\sigma, \epsilon)\pi(\sigma, \epsilon,t) + \lambda(\sigma, -\epsilon)\pi(\sigma, -\epsilon, t)
$$
- Outgoing Flow: $-[\sum_j w_j(\sigma, \epsilon)] p(\sigma, \epsilon, t)$ and means that the system flips to move out of the configuration.

- Incoming Flow: $\sum_j w_j(F_j\sigma, \epsilon)p(F_j\sigma, \epsilon, t)$ is the probability that the system is in a neighbouring state and flips into the current state

- Forward Flip Rate: $\lambda(\sigma, \epsilon)\pi(\sigma, \epsilon,t)$

- Backward Flip Rate: $\lambda(\sigma, -\epsilon)\pi(\sigma, -\epsilon, t)$

Here is the expectation of the observable which allow us to understand how the state evolves

$$
E[A(t)] = \sum_{\epsilon = \pm 1}\sum_{\sigma} A(\sigma, \epsilon) \pi(\sigma, \epsilon, t)
$$

- $A(t)$ and $A(\sigma, \epsilon)$ refers to the average and the actual value of the observable. Here it is Magnetization: $A(\sigma, \epsilon) = m(\sigma) = \frac{1}{N}\sum_j\sigma_j$

$$
\frac{d}{dt}E[m(t)] = \sum_{\epsilon = \pm 1}\sum_{\sigma} A(\sigma, \epsilon) \frac{d}{dt}\pi(\sigma, \epsilon, t)
$$



### Deriving Magnetization Equation
Firstly, our $A(t)$ and $A(\sigma, \epsilon)$ will be replaced by $m(t)$ and $m(\sigma, \epsilon)$ since we are using magnetization

We can substitute the master equation into the expectation equation to get this

$$
\frac{d}{dt}E[m(t)] = \sum_{\epsilon = \pm 1}\sum_{\sigma} m(\sigma, \epsilon)(-[\sum_j w_j(\sigma, \epsilon)] \pi(\sigma, \epsilon, t) + \sum_j w_j(F_j\sigma, \epsilon)\pi(F_j\sigma, \epsilon, t) - \lambda(\sigma, \epsilon)\pi(\sigma, -\epsilon,t) + \lambda(\sigma, -\epsilon)\pi(\sigma, -\epsilon, t))
$$

rearranging a bit we get this

$$
\frac{d}{dt}E[m(t)] = \sum_{\epsilon = \pm 1}\sum_{\sigma} -m(\sigma, \epsilon)(\lambda(\sigma, \epsilon)\pi(\sigma, -\epsilon,t)) + m(\sigma, \epsilon)(\lambda(\sigma, -\epsilon)\pi(\sigma, -\epsilon, t)) + m(\sigma, \epsilon)(-[\sum_j w_j(\sigma, \epsilon)] \pi(\sigma, \epsilon, t) + \sum_j w_j(F_j\sigma, \epsilon)\pi(F_j\sigma, \epsilon, t))
$$

We can cancel the $\lambda(\sigma, \epsilon)\pi(\sigma, -\epsilon,t) + \lambda(\sigma, -\epsilon)\pi(\sigma, -\epsilon, t)$ component because $m(\sigma)$ is solely determined by the spin configuration. 

So $m(\sigma, \epsilon) = m(\sigma, -\epsilon)$ and the 2 cancel out. This is the equation we are left with

$$
\frac{d}{dt}E[m(t)] = \sum_{\epsilon = \pm 1}\sum_{\sigma}m(\sigma, \epsilon)(-[\sum_j w_j(\sigma, \epsilon)] \pi(\sigma, \epsilon, t) + \sum_j w_j(F_j\sigma, \epsilon)\pi(F_j\sigma, \epsilon, t))
$$

Next, we do some rearranging

$$
\frac{d}{dt}E[m(t)] = \sum_{\epsilon, \sigma}m(\sigma, \epsilon)(-[\sum_j w_j(\sigma, \epsilon)] \pi(\sigma, \epsilon, t) + \sum_j w_j(F_j\sigma, \epsilon)\pi(F_j\sigma, \epsilon, t))
$$

$$
\frac{d}{dt}E[m(t)] = \sum_{\epsilon, \sigma}(-[\sum_j m(\sigma, \epsilon)w_j(\sigma, \epsilon)] \pi(\sigma, \epsilon, t) + \sum_j m(\sigma, \epsilon)w_j(F_j\sigma, \epsilon)\pi(F_j\sigma, \epsilon, t))
$$

Then lets relabel the $\sigma' = F_j\sigma$. Also notice that 
$$
\sum_{x \in (a, a')} x*f(x) + x'*g(x') = \sum_{x \in (a, a')} x*f(x) + x*g(x) = x\sum_{x \in (a, a')} f(x) + g(x)
$$

and therefore

$$
\frac{d}{dt}E[m(t)] = \sum_{\epsilon, \sigma}\sum_j (-m(\sigma, \epsilon)w_j(\sigma, \epsilon) \pi(\sigma, \epsilon, t) + m(F_j\sigma, \epsilon)w_j(\sigma, \epsilon)\pi(\sigma, \epsilon, t))
$$
It is important to know that this only works for ising models.

take the probability out (doesn't have a j term) and factor $w_j$
$$
\frac{d}{dt}E[m(t)] = \sum_{\epsilon, \sigma}\pi(\sigma, \epsilon, t) \sum_j w_j(\sigma, \epsilon)(-m(\sigma, \epsilon) + m(F_j\sigma, \epsilon))
$$

notice that the right side is just in the form of $\sum p(x) f(x)$, so we can represent it as $E[f(x)]$

$$
\frac{d}{dt}E[m(t)] = E[\sum_j w_j(\sigma, \epsilon)(-m(\sigma, \epsilon) + m(F_j\sigma, \epsilon))]
$$

substitute gaubler transition probability and local change in magnetization $m(F_j\sigma) - m(\sigma) = -\frac{2}{N}\sigma_j$


$$
\frac{d}{dt}E[m(t)] = E[\sum_j \frac{1}{2}\alpha[1 - \frac{1}{2}\gamma\sigma_j(\sigma_{j-1} + \sigma_{j + 1})](1 - \delta\epsilon\sigma_j)(-\frac{2}{N}\sigma_j)]
$$
$$
\frac{d}{dt}E[m(t)] = E[\frac{1}{N} \sum_j \frac{1}{2}\alpha[1 - \frac{1}{2}\gamma\sigma_j(\sigma_{j-1} + \sigma_{j + 1})](1 - \delta\epsilon\sigma_j)(-2\sigma_j)]
$$

expanding and simplifying with $\sigma_j^2 = 1$, we get

$$
\frac{d}{dt}E[m(t)] = E[\frac{1}{N} \sum_j -\sigma_j + \delta\epsilon + \frac{1}{2}\gamma(\sigma_{j-1} + \sigma_{j + 1}) - \frac{1}{2}\delta\epsilon\gamma\sigma_j(\sigma_{j-1} + \sigma_{j + 1})]
$$


use the following to simplify further
$$
1/N \sum \sigma_j = m(\sigma)
$$
$$
1/N \sum \delta\epsilon = \delta\epsilon
$$
$$
1/N \sum \frac{1}{2}\gamma(\sigma_{j-1} + \sigma_{j + 1}) = \gamma m(\sigma)
$$
$$
1/N \sum \frac{1}{2}\delta\epsilon\gamma\sigma_j(\sigma_{j-1} + \sigma_{j + 1}) = -\delta\gamma\epsilon u(\sigma), u=local\_energy\_correlation
$$

Final Equation
$$
\frac{1}{\alpha}\frac{d}{dt}E[m(t)] = -(1 - \gamma)(E[m(t)]) + \delta E[\epsilon(t)] - \delta \gamma E[\epsilon u(t)]
$$

### Deriving the Direction Equation

This is very similar to the derivation of the magnetization.

Here, the observable is the lifting variable. So similarly $A(t)$ and $A(\sigma, \epsilon)$ will be equal to $\epsilon$

$$
\frac{d}{dt}E[\epsilon(t)] = \sum_{\epsilon = \pm 1}\sum_{\sigma} \epsilon(\sigma, \epsilon)(-[\sum_j w_j(\sigma, \epsilon)] \pi(\sigma, \epsilon, t) + \sum_j w_j(F_j\sigma, \epsilon)\pi(F_j\sigma, \epsilon, t) - \lambda(\sigma, \epsilon)\pi(\sigma, -\epsilon,t) + \lambda(\sigma, -\epsilon)\pi(\sigma, -\epsilon, t))
$$

However, now instead, the the spin flip part cancels

$$
\frac{d}{dt}E[\epsilon(t)] = \sum_{\epsilon, \sigma}\epsilon(-\lambda(\sigma, \epsilon)\pi(\sigma, -\epsilon,t) + \lambda(\sigma, -\epsilon)\pi(\sigma, -\epsilon, t))
$$

rearrange again

$$
\frac{d}{dt}E[\epsilon(t)] = \sum_{\epsilon, \sigma}-\epsilon\lambda(\sigma, \epsilon)\pi(\sigma, -\epsilon,t) + \epsilon \lambda(\sigma, -\epsilon)\pi(\sigma, -\epsilon, t)
$$

Doing the same summation trick, we can relabel $\epsilon \rightarrow -\epsilon$ 

$$
\frac{d}{dt}E[\epsilon(t)] = \sum_{\epsilon, \sigma}-\epsilon\lambda(\sigma, \epsilon)\pi(\sigma, -\epsilon,t) + -\epsilon \lambda(\sigma, \epsilon)\pi(\sigma, \epsilon, t)
$$

$$
\frac{d}{dt}E[\epsilon(t)] = \sum_{\epsilon, \sigma}-2\epsilon\lambda(\sigma, \epsilon)\pi(\sigma, -\epsilon,t)
$$

equate it to expectation

$$
\frac{d}{dt}E[\epsilon(t)] = E[-2\epsilon\lambda(\sigma, \epsilon)]
$$

substitute transition probability

$$
\frac{d}{dt}E[\epsilon(t)] = E[-2\epsilon \frac{1}{2}\alpha N [1 - \gamma u(\sigma) + \delta (1 - \gamma) \epsilon m(\sigma)]
]
$$

since $\epsilon^2 = 1$ we can simplify to get

$$
\frac{1}{\alpha} \frac{d}{dt} E[\epsilon(t)] = -\delta(1 - \gamma) N E[m(t)] - N E[\epsilon(t)] + \gamma N E[\epsilon u(t)] 
$$


### Deriving the Relaxation time

**Magnetization Equation**
$$
\frac{1}{\alpha}\frac{d}{dt}E[m(t)] = -(1 - \gamma)(E[m(t)]) + \delta E[\epsilon(t)] - \delta \gamma E[\epsilon u(t)]
$$
**Direction Equation**
$$
\frac{1}{\alpha} \frac{d}{dt} E[\epsilon(t)] = -\delta(1 - \gamma) N E[m(t)] - N E[\epsilon(t)] + \gamma N E[\epsilon u(t)] 
$$

from the direction equation, when N becomes very large and we can equate both sides
$$
\frac{1}{\alpha} \frac{d}{dt} E[\epsilon(t)] / N = -\delta(1 - \gamma) E[m(t)] - E[\epsilon(t)] + \gamma E[\epsilon u(t)] 
$$
$$
\lim_{n \rightarrow \inf}
$$

$$
0 = -\delta(1 - \gamma) E[m(t)] - E[\epsilon(t)] + \gamma E[\epsilon u(t)] 
$$
$$
E[\epsilon(t)] - \gamma E[\epsilon u(t)]  = -\delta(1 - \gamma) E[m(t)]
$$

Notice how the second term in the right part of the magnetization equation is equal to that and therefore we can substitute
$$
\frac{1}{\alpha}\frac{d}{dt}E[m(t)] = -(1 - \gamma)(E[m(t)]) + -\delta\delta(1 - \gamma) E[m(t)]
$$

After some simplifying
$$
\frac{1}{\alpha}\frac{d}{dt}E[m(t)] = -(1 + \delta^2)(1 - \gamma) E[m(t)]
$$

Based on the exponential decay of Magnetization density.
$$
E[m(t)] = E[m(0)]e^{-\frac{t}{\tau}}
$$
we can substitute and get

$$
\frac{1}{\alpha}\frac{d}{dt}E[m(0)]e^{-\frac{t}{\tau}} = -(1 + \delta^2)(1 - \gamma) E[m(0)]e^{-\frac{t}{\tau}}
$$

$$
-\frac{1}{\tau}E[m(0)]e^{-\frac{t}{\tau}} = -\alpha(1 + \delta^2)(1 - \gamma) E[m(0)]e^{-\frac{t}{\tau}}
$$

from this we can finally see that 
$$
-\frac{1}{\tau} = -\alpha(1 + \delta^2)(1 - \gamma) 
$$

and therefore

$$
{\tau} = \frac{1}{\alpha(1 + \delta^2)(1 - \gamma)}
$$

since the formula for a reversible markov chain is 

$$
\tau_{rev} = \frac{1}{\alpha(1 - \gamma)}
$$

with a max skew of $\delta = 1$
$$
\tau = \frac{1}{\alpha(1 + \delta^2)(1 - \gamma)} = \frac{1}{2\alpha(1 - \gamma)}
$$

and therefore converges twice as fast
